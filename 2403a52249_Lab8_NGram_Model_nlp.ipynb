**STEP 2:import libraries**

```python
# Library for text processing (regular expressions)
import re

# Counter helps to count word frequencies and N-grams
from collections import Counter

# NLTK for tokenization and n-gram generation
import nltk
from nltk.util import ngrams

# Math library for probability and perplexity calculations
import math
```

**STEP 3: Load Dataset**

```python
# Given corpus
corpus = [
"Students should revise lessons daily",
"Regular practice improves learning skills",
"Time management helps in exam preparation",
"Group study increases subject understanding",
"Online resources support self learning",
"Note taking helps in quick revision",
"Research work develops critical thinking",
"Library study provides a quiet environment",
"Proper planning reduces academic stress",
"Consistent study leads to better performance"
]

# Display sample text
print("Sample Text:")
for line in corpus[:10]:
    print(line)
```

**STEP 4:Preprocessing**

```python
# Lowercase and remove punctuation
clean_corpus = []
for line in corpus:
    line = line.lower()
    line = re.sub(r'[^a-z\s]', '', line)
    clean_corpus.append(line)

# Tokenization
tokens = []
for line in clean_corpus:
    tokens.extend(line.split())

print("Tokens:", tokens)
```

**STEP 5:Identify rare words[UNK]**

```python
# Count word frequency
word_counts = Counter(tokens)

# Rare words = frequency = 1
rare_words = {word for word, count in word_counts.items() if count == 1}

# Replace rare words with UNK
tokens_unk = [word if word not in rare_words else "UNK" for word in tokens]

print("Rare words:", rare_words)
```

**deployment of bi-gram model**

```python
def predict_next_word_bigram(text, bigram_counts, unigram_counts, vocab_size):
    words = text.lower().split()

    if not words:
        return "<NO_CONTEXT>"

    prefix_word = words[-1]

    max_prob = -1
    predicted_word = None

    prefix_word_count = unigram_counts.get(prefix_word, 0)

    for next_word in unigram_counts.keys():
        current_bigram = (prefix_word, next_word)
        bigram_count = bigram_counts.get(current_bigram, 0)

        denominator = prefix_word_count + vocab_size
        if denominator == 0:
            prob = 0
        else:
            prob = (bigram_count + 1) / denominator

        if prob > max_prob:
            max_prob = prob
            predicted_word = next_word

    return predicted_word

ip_text=input("enter text")
# Call the function and get the predicted next word
next_word1 = predict_next_word_bigram(ip_text, bigram_counts, unigram_counts, V)

# Print the predicted word
print(f"Given sequence: '{ip_text}', predicted next word: '{next_word1}'")
```

**Deployment of Add-K Smoothening based Tri-Gram Model**

```python
def predict_next_word_trigram_K(text, trigram_counts, bigram_counts, k, vocab_size, unigram_vocabulary):
    # Preprocess the input text
    words = text.lower().split()

    if len(words) < 2:
        # If there are not enough words to form a bigram prefix, cannot predict next word with trigram model
        return "<NOT_ENOUGH_CONTEXT>"

    # Get the last two words as the prefix for prediction
    w1, w2 = words[-2], words[-1]

    max_prob = -1
    predicted_word = None

    # Iterate over all possible words in the vocabulary for the next word
    for next_word in unigram_vocabulary.keys():
        current_trigram = (w1, w2, next_word)
        current_bigram_prefix = (w1, w2)

        # Get counts for the trigram and the bigram prefix
        trigram_count = trigram_counts.get(current_trigram, 0)
        bigram_prefix_count = bigram_counts.get(current_bigram_prefix, 0)

        # Calculate the probability using Add-K smoothing
        # P(next_word | w1, w2) = (C(w1, w2, next_word) + k) / (C(w1, w2) + k * V)
        denominator = bigram_prefix_count + k * vocab_size
        if denominator == 0: # Avoid division by zero if bigram_prefix_count is 0 and k*vocab_size is 0
            prob = 0
        else:
            prob = (trigram_count + k) / denominator

        if prob > max_prob:
            max_prob = prob
            predicted_word = next_word

    return predicted_word

ip_text=input("enter text")
# Call the function with the defined Trigrams_counts, bigram_counts, smoothing parameter (0.5),
# vocabulary size (V), and unigram_counts (for vocabulary access).
next_word1 = predict_next_word_trigram_K(ip_text, Trigrams_counts, bigram_counts, 0.5, V, unigram_counts)
print(f"Given sequence: '{ip_text}', predicted next word: '{next_word1}'")
```

**STEP 8:Perplexity Calculation**

test sentence

```python
test_sentences = [
"students revise daily",
"time management helps students",
"group study improves learning",
"online learning improves skills",
"research improves understanding"
]
```

perplexity function

```python
def perplexity(sentence, unigram_counts):
    words = sentence.lower().split()
    N = len(words)
    log_prob = 0

    for word in words:
        prob = (unigram_counts[word] + 1) / (sum(unigram_counts.values()) + vocab_size)
        log_prob += math.log(prob)

    return math.exp(-log_prob / N)

# Calculate perplexity
print("Perplexity Results:")
for sent in test_sentences:
    print(sent, ":", perplexity(sent, unigrams))
```
